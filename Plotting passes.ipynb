# Plotting passes

```python
import matplotlib.pyplot as plt
import numpy as np
from mplsoccer import Pitch, Sbopen
```

```python
parser = Sbopen()
df, related, freeze, tactics = parser.event(69303)
passes = df.loc[df['type_name'] == 'Pass'].loc[df['sub_type_name'] != 'Throw-in'].set_index('id')
```

```python
passes.head()
```

```python
passes[['team_name', 'player_name']].head(40)
```

```python
#drawing pitch
pitch = Pitch(line_color = "black")
fig, ax = pitch.draw(figsize=(10, 7))

for i,pass1 in passes.iterrows():
    
    if pass1['player_name']=='AndrÃ©s Iniesta LujÃ¡n':
        x=pass1['x']
        y=pass1['y']
        #plot circle
        passCircle=plt.Circle((x,y),2,color="red")
        passCircle.set_alpha(.2)
        ax.add_patch(passCircle)
        dx=pass1['end_x']-x
        dy=pass1['end_y']-y
        #plot arrow
        passArrow=plt.Arrow(x,y,dx,dy,width=3,color="blue")
        ax.add_patch(passArrow)

ax.set_title("Iniesta passes", fontsize = 24)
fig.set_size_inches(10, 7)
plt.show()
```

```python

#drawing pitch
pitch = Pitch(line_color = "black")
fig1, ax1 = pitch.draw(figsize=(10, 7))

for i,pass1 in passes.iterrows():
    
    if pass1['player_name']=='Daniel Alves da Silva':
        x=pass1['x']
        y=pass1['y']
        #plot circle
        passCircle=plt.Circle((x,y),2,color="red")
        passCircle.set_alpha(.2)
        ax1.add_patch(passCircle)
        dx=pass1['end_x']-x
        dy=pass1['end_y']-y
        #plot arrow
        passArrow=plt.Arrow(x,y,dx,dy,width=2,color="blue")
        ax1.add_patch(passArrow)

ax1.set_title("Dani Alves passes", fontsize = 24)
fig1.set_size_inches(10, 7)
plt.show()
```

## Plotting multiple pass maps on one figure

```python
passBarca = (df.type_name == 'Pass') & (df.team_name == 'Barcelona') & (df.sub_type_name != 'Throw-in')
df_passes = df.loc[passBarca, ['x','y','end_x', 'end_y', 'player_name']]

names = df_passes['player_name'].unique()
```

```python
pitch = Pitch(line_color='black', pad_top=20)
fig2,ax2 = pitch.grid(ncols=4, nrows=4, grid_height=0.85, title_height=0.06, axis=False, endnote_height=0.04, title_space=0.04, endnote_space=0.01)

for name, ax in zip(names, ax2['pitch'].flat[:len(names)]):
    ax.text(60,-10, name, ha='center', va='center', fontsize=8)
    
    df_player = df_passes.loc[df_passes['player_name']== name]
    
    pitch.scatter(df_player.x, df_player.y, alpha = 0.2, s=50, color='blue', ax=ax)
    pitch.arrows(df_player.x, df_player.y, df_player.end_x, df_player.end_y, color='blue', ax=ax, width=1)
    
for ax in ax2['pitch'][-1,16 - len(names):]:
    ax.remove()
    
ax2['title'].text(0.5,0.5,'Barcelona passes against Getafe', ha='center', va='center', fontsize=30)
plt.show()
    
```
